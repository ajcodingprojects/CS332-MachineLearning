```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ames_housing = pd.read_csv("ames_housing_no_missing.csv")
ames_housing.head()
```

```python
ames_housing.describe()
```

```python
features_of_interest = [
    "LotFrontage",
    "LotArea",
    "PoolArea",
    "YearBuilt",
    "YrSold",
]
target_name = "SalePrice"

# separate into data and target
data, target = (
    ames_housing[features_of_interest],
    ames_housing[target_name],
)
```

```python
sns.pairplot( ames_housing, x_vars=[
"LotFrontage","LotArea","PoolArea","YearBuilt","YrSold"], y_vars='SalePrice', kind=
'reg', plot_kws={'line_kws':{'color':'teal'}})
```

```python
linear_regression = make_pipeline(
    PolynomialFeatures(degree=2, include_bias=False), LinearRegression()
    ).set_output(transform="pandas")
cv_results = cross_validate(
    linear_regression,
    data,
    target,
    cv=10,
    scoring="neg_mean_squared_error",
    return_train_score=True,
    return_estimator=True,
)
```

```python
print(cv_results["train_score"])
print(cv_results["test_score"])
```

```python
train_error = -cv_results["train_score"]
print(
"Mean squared error of linear regression model on the train set:\n"
f"{train_error.mean():.2e} Â± {train_error.std():.2e}"
)
```

```python
test_error = -cv_results["test_score"]
print(
"Mean squared error of linear regression model on the test set:\n"
f"{test_error.mean():.2e} Â± {test_error.std():.2e}"
)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
# Step 1: Define the pipeline
pipeline = Pipeline([
('scaler', StandardScaler()), # Preprocessing step (feature scaling)
('poly_features', PolynomialFeatures(degree=2, include_bias=False)),
('ridge', Ridge(random_state= 42)) # Regression step
])
# Step 2: Set up the parameter grid
param_grid = {
'ridge__alpha': [0.1, 1, 10, 100, 1000] # Adjust the grid values as needed
}
# Step 3: Initialize GridSearchCV
grid_search = GridSearchCV(
estimator=pipeline,
param_grid=param_grid,
scoring='neg_mean_squared_error', # You can choose another scoring if needed
cv=KFold(n_splits=5, shuffle=True, random_state=42), # Cross-validation strategy
verbose=1, # For detailed logging information (use 0 for no verbosity)
)
# Step 4: Fit GridSearchCV to the data
grid_search.fit(data, target)
# Best parameters found
print("Best parameters found: ", grid_search.best_params_)
# Best score found
print("Best score found: ", -grid_search.best_score_) # Change sign because scoring isnegative MSE
# You can also retrieve the best estimator directly
best_model = grid_search.best_estimator_
```

```python
from sklearn.linear_model import LassoCV
# Define a range of alphas for Lasso
alphas = np.logspace(-7, 5, num=100)
# Create a pipeline that includes scaling, polynomial feature expansion, and Lassoregression
lasso = make_pipeline(
MinMaxScaler(),
PolynomialFeatures(degree=2, include_bias=False),
LassoCV(alphas=alphas, cv=5, max_iter=10000),
)
# Define the cross-validation strategy
cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)
# Execute cross-validation
cv_results = cross_validate(
lasso,
data,
target,
cv=cv,
scoring="neg_mean_squared_error",
return_train_score=True,
return_estimator=True
)
```

```python
train_error = -cv_results["train_score"]
print(
"Mean squared error of tuned lasso model on the train set:\n"
f"{train_error.mean():.2e} Â± {train_error.std():.2e}"
)
test_error = -cv_results["test_score"]
print(
"Mean squared error of tuned lasso model on the test set:\n"
f"{test_error.mean():.2e} Â± {test_error.std():.2e}"
)
```

```python
mse_alphas = [
est[-1].mse_path_.mean(axis=1) for est in cv_results["estimator"]
]
# est[-1] accesses the LassoCV object in the pipeline
# mse_path_ contains the mean square errors for each fold and each alpha
# We take the mean of mse_path_ across all folds (axis=1)
# Convert list of mean squared errors to a DataFrame
# Each row in the DataFrame corresponds to an alpha value
# Each column corresponds to a different estimator
cv_alphas = pd.DataFrame(mse_alphas, columns=alphas)
# Calculate the mean and standard deviation of the mean squared errors across estimatorsfor each alpha
cv_alphas = cv_alphas.aggregate(["mean", "std"]).T
cv_alphas
```

```python
from sklearn.linear_model import ElasticNet
# Define the pipeline
elasticnet_pipeline = Pipeline([
('scaler', MinMaxScaler()),
('poly_features', PolynomialFeatures(degree=2, include_bias=False)),
('elasticnet', ElasticNet()) # Initially, don't set alpha or l1_ratio here
])
# Set up the parameter grid
param_grid = {
'elasticnet__alpha': [0.001, 0.01, 0.1, 1, 10, 100],
'elasticnet__l1_ratio': [0.1, 0.5, 0.7, 0.9, 0.95, 1] # Values between 0 and 1
}
# Initialize GridSearchCV
grid_search = GridSearchCV(
estimator=elasticnet_pipeline,
param_grid=param_grid,
scoring='neg_mean_squared_error', # or any other appropriate scoring method
cv=KFold(n_splits=5, shuffle=True, random_state=42), # Define your cross-validation
verbose=1, # For detailed logging information; use 0 or 1 typically
n_jobs=-1 # Use all available cores on your machine
)
# Fit GridSearchCV to the data
grid_search.fit(data, target)
# Best parameters found
print("Best parameters found: ", grid_search.best_params_)
# Best score found
print("Best score found: ", -grid_search.best_score_) # Scoring is neg mean squarederror, hence the negation
# Best estimator
best_model = grid_search.best_estimator_
```
