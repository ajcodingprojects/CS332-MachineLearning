```python
import statistics
import matplotlib.pyplot as plt

class Person:
    name = ''
    sex = ''
    occurrences = 0
    
    def __init__(self, n, s, o):
        self.name = n
        self.sex = s
        self.occurrences = o
        

def count_occurrences_in_file(path, target):
    '''Counts the occurences of a baby being born with a certain name in a certain file
    Parameters
    ----------
        path: the file path that contains data stored as a csv with 3 values:
            name, gender, occurences
        target: the target name to find
    Returns
    ----------
    A list, first value the number found, the second number the total babies born'''
    with open(path, 'r') as file:
        # Create vars to keep track of totals
        total = 0
        found = 0
        # Iterate over the lines of the file
        for line in file:
            # find the occurences
            l = line.split(',')
            person = Person(l[0], l[1], l[2])
            total = total + int(person.occurrences)
            if (person.name.__eq__(target)):
                found = found + int(person.occurrences)
            
        return [found, total]
    
def count_occurrences_in_years(years, target):
    '''Counts the occurences of a baby being born with a certain name in a range of years
    Parameters
    ----------
        years: the range of years
        target: the target name to find
    Returns
    ----------
    A list containing lists for each year with the first value the number found, the second number the total babies born'''

    yearly_report = []
    for year in years:
        yearly_report.append(count_occurrences_in_file('names/yob'+str(year)+'.txt', target))
    
    return yearly_report
        
    
def plot_national_occurrences(data, years, target):
    '''Plots the number of occurrences in the given range of years
    Parameters
    ----------
        data: the data of occurrences to plot
        years: the year range of the data
        target: the target name of the data
    Returns
    -------
    void'''
    
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    
    occurrences = []
    
    for d in data:
        occurrences.append(d[0])
        
    l1 = ax.plot(years, occurrences, 'b-')
    ax.set_title('Number of Occurrences for '+ str(target))
    ax.set_xlabel('Years')
    ax.set_ylabel('Number of Occurrences')
    plt.show()
    
    return
    
    
def plot_national_percentages(data, years, target):
    '''Plots the percentages of occurrences of the target name in the given year range
    Parameters
    ----------
        data: the data of occurrences to plot
        years: the year range of the data
        target: the target name of the data
    Returns
    -------
    void'''
    
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    
    percentages = []
    
    for d in data:
        percentages.append((d[0]/d[1])*100)
        
    l1 = ax.plot(years, percentages, 'b-')
    ax.set_title('Percentages of Occurrences for '+ str(target))
    ax.set_xlabel('Years')
    ax.set_ylabel('Percentage of Occurrences')
    plt.show()
    
    return
    
    
```

```python
years = range(1880, 2022);
name = 'Andrew'
occurrences = count_occurrences_in_years(years, name)

plot_national_occurrences(occurrences, years, name)

```

```python
years = range(1880, 2022);
name = 'Andrew'
occurrences = count_occurrences_in_years(years, name)

plot_national_percentages(occurrences, years, name)
```
