# Pytorch

```python
import torch
import torch.nn as nn
from torch.autograd import Variable
```

```python
penguins = pd.read_csv("penguins_regression.csv")
penguins.head()
```

```python
X = penguins.drop(["Body Mass (g)"], axis=1)
y = penguins[["Body Mass (g)"]]
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#For X:
sc_x = StandardScaler()
x_train_scaled = sc_x.fit_transform(x_train)
x_test_scaled = sc_x.transform(x_test)

#For Y:
sc_y = StandardScaler()
y_train_scaled = sc_y.fit_transform(y_train)
y_test_scaled = sc_y.transform(y_test)
```

```python
x_train_scaled = np.array(x_train_scaled,dtype=np.float32)
y_train_scaled = np.array(y_train_scaled,dtype=np.float32)
# Convert inputs and targets to tensors
inputs = torch.from_numpy(x_train_scaled)
targets = torch.from_numpy(y_train_scaled)
```

```python
# create class
class LinearRegression(nn.Module):
    def __init__(self,input_dim,output_dim):
        # super function. It inherits from nn.Module and we can access everythink in nn.Module
        super(LinearRegression,self).__init__()
        # Linear function.
        self.linear = nn.Linear(input_dim,output_dim)
    def forward(self,x): # x:inputs
        return self.linear(x)
```

```python
inputDim = 1 # takes variable 'x'
outputDim = 1 # takes variable 'y'
learningRate = 0.01
epochs = 100
model = LinearRegression(inputDim, outputDim)
model
```

```python
# MSE
mse = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)
```

```python
loss_list=[] #only needed to show how the error descreases with the epochs
for epoch in range(epochs):
    # Clear gradient buffers because we don't want any gradient from previous epoch tocarry forward,
    # dont want to cummulate gradients
    optimizer.zero_grad()
    # get output from the model, given the inputs
    results = model(inputs)
    # get loss for the predicted output
    # Calculate Loss
    loss = mse(results, targets)
    print(loss)
    # get gradients w.r.t to parameters
    loss.backward()
    # update parameters
    optimizer.step()
    # store loss
    loss_list.append(loss.data)
    #optional print
    print('epoch {}, loss {}'.format(epoch, loss.item()))
```

```python
plt.plot(range(epochs),loss_list)
plt.xlabel("Number of Iterations")
plt.ylabel("Loss")
plt.show()
```

```python
input_x_test = torch.from_numpy(x_test_scaled)
predicted = model(input_x_test.float()).data.numpy()
#print(predicted)#optional
```

```python
plt.clf()
plt.plot(x_train_scaled, y_train_scaled, 'go', label='True data', alpha=0.5)
plt.plot(x_test_scaled, predicted, '--', label='Predictions', alpha=0.5)
plt.legend(loc='best')
plt.show()
model_error = mean_absolute_error(y_test_scaled, predicted)
print(f"The mean absolute error of the optimal model is {model_error:.2f} for thestandardized scale")
model_error = mean_squared_error(y_test_scaled, predicted)
print(f"The mean squared error of the optimal model is {model_error:.2f} for thestandardized scale")
```
